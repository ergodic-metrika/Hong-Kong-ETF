```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import matplotlib.dates as mdates
```

```python
# Load the Excel file
excel_file = pd.ExcelFile('D:\Derivatives Trading\TAIEX derivatives trading record.xlsx')
```

```python
#########################################################
#ETF Portfolio Allocation
sheet_name_2 = 'ETF_Allocation' 
df_2 = excel_file.parse(sheet_name_2)
# Replace infinite values with NaN, then drop rows with NaN in critical columns
df_2.replace([np.inf, -np.inf], np.nan, inplace=True)
df_2.dropna(subset=['Total', 'Percentage', 'ETF'], inplace=True)

ETF=df_2["ETF"]
df_2['Total'] = df_2['Shares'] * df_2['Value']  
total_value = df_2['Total'].sum() 
df_2['Percentage'] = (df_2['Total'] / total_value) * 100
```

```python
# Custom colors for the pie chart slices
colors = ['lightpink','violet','hotpink','tan','gold','lightskyblue', 'deepskyblue', 'cornflowerblue','royalblue', 'powderblue','aquamarine','turquoise']

# Plotting the pie chart
plt.figure(figsize=(10, 8))  # Set the figure size
plt.pie(df_2['Percentage'], labels=df_2['ETF'], colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Taiwan High Dividend ETF Allocation', fontweight='bold', fontsize=14)  
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Show the plot
plt.show()
```
