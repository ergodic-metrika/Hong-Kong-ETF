```python
#Package
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy.optimize import minimize
import matplotlib.pyplot as plt
```

```python
# Read excel file
file_path = r'D:\Derivatives Trading\BITX QQQ TSLA.xlsx'
df = pd.read_excel(file_path, index_col='Date', parse_dates=True)

# Calculate daily returns
returns = df.pct_change()

# Remove any NaN values
returns = returns.dropna()
```

```python
# Regression 1: TSLA vs BITX
X1 = sm.add_constant(returns['TSLA'])
model1 = sm.OLS(returns['BITX'], X1).fit()
print("\nRegression Results: TSLA vs BITX")
print("Dependent Variable: BITX")
print("Independent Variable: TSLA")
print(model1.summary())

# Scatter plot for TSLA vs BITX
plt.figure(figsize=(10, 6))
plt.scatter(returns['TSLA'], returns['BITX'], alpha=0.5)
plt.plot(returns['TSLA'], model1.predict(X1), color='red', linewidth=2)
plt.xlabel('TSLA Returns')
plt.ylabel('BITX Returns')
plt.title('TSLA vs BITX Returns with Regression Line')
plt.grid(True)
plt.show()
```

```python
# Regression 2: QQQ vs BITX
X2 = sm.add_constant(returns['QQQ'])
model2 = sm.OLS(returns['BITX'], X2).fit()
print("\nRegression Results: QQQ vs BITX")
print("Dependent Variable: BITX")
print("Independent Variable: QQQ")
print(model2.summary())

# Scatter plot for QQQ vs BITX
plt.figure(figsize=(10, 6))
plt.scatter(returns['QQQ'], returns['BITX'], alpha=0.5)
plt.plot(returns['QQQ'], model2.predict(X2), color='red', linewidth=2)
plt.xlabel('QQQ Returns')
plt.ylabel('BITX Returns')
plt.title('QQQ vs BITX Returns with Regression Line')
plt.grid(True)
plt.show()
```
